```python
import matplotlib.pyplot as plt  # 导入绘图库

# 数据点
x = [1, 2, 3, 4, 5]  # 定义数据点的 x 值
y1 = [1, 4, 9, 16, 25]  # 定义第一条曲线的 y 值 (y = x^2)
y2 = [25, 20, 15, 10, 5]  # 定义第二条曲线的 y 值 (y = -x^2 + 26)

# 绘制两条曲线并添加标签
plt.plot(x, y1, label='y = x^2')  # 绘制第一条曲线并添加标签
plt.plot(x, y2, label='y = -x^2 + 26')  # 绘制第二条曲线并添加标签

# 向图表中添加图例
plt.legend()  # 显示带有曲线标签的图例

# 显示图表
plt.show()  # 渲染并显示图表
```

```python
import matplotlib.pyplot as plt

# 创建数据
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 20, 15, 10, 5]

# 绘制两条曲线，不添加标签
plt.plot(x, y1)
plt.plot(x, y2)

# 不调用 plt.legend()，所以没有图例

# 显示图表
plt.show()
```
